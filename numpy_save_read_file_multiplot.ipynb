### create multidimensional arrary

```python
t = np.arange(1, 5.0, 0.1)               #create one dimensional array, (start, end, interval)
#print(t)
#print(type(t))
data1 = np.exp(t)                         #array of exponential function
data2 = np.sin(2 * np.pi * t)             #array of exponential function
data3 = np.cos(2 * np.pi * t)

sqdata1 = np.exp(t)**2                   #square of above functions
sqdata2 =np.sin(2 * np.pi * t)**2
sqdata3 = np.cos(2 * np.pi * t)**2

#print(t.dtype)
#print(t)
type(data1)

# convert/transform row into column
#
tt = t.reshape(-1,1)                  #convert one dimensional array into column
#print(t)
#print(tt)

tdata1 = data1.reshape(-1,1)
#print(data1)
#print(tdata1)

tdata2 = data2.reshape(-1,1)
#print(data2)
#print(tdata2)

tdata3 = data3.reshape(-1,1)
#print(data3)
#print(tdata3)

tsqdata1 = sqdata1.reshape(-1,1)
#print(tsqdata1)
tsqdata2 = sqdata2.reshape(-1,1)
tsqdata3 = sqdata3.reshape(-1,1)

arexp = np.concatenate((tt,tdata1,tsqdata1),axis = 1)
arsin = np.concatenate((tt,tdata2,tsqdata2),axis = 1)
arcos = np.concatenate((tt,tdata3,tsqdata3),axis = 1)
print(arexp)
# OR using column_stack
#arexp = np.column_stack([tt,tdata1,tsqdata1])
#arsin = np.column_stack([tt,tdata2,tsqdata2])
#arcos = np.concatenate([tt,tdata3,tsqdata3])
```

## plot the data with multiple vertical axes
### plot 1: simple plot 

```python
f_size = 20   #font size
l_width = 3  #line width

fig, plot1 = plt.subplots(figsize=(7,5)) #this is host plot
plt.xticks(fontsize=f_size)
plt.yticks(fontsize=f_size)
#plot1.set_title('Title: Overlay plots',fontsize=f_size)

p1 = plot1.plot(r_te, r_exp, color="red", linewidth = l_width, label = 'Exponential')
p2 = plot1.plot(r_ts, r_sin, color="blue", linewidth = l_width, label = 'Sine')
p3 = plot1.plot(r_tc, r_cos, color="black", linewidth = l_width, linestyle='--',marker='o', label = 'cosine')

plot1.set_xlabel('Time (s)', fontsize = f_size)
plot1.set_ylabel('Functions',  fontsize = f_size)
plot1.tick_params(axis='both', direction='out')   #direction = 'in', 'out', 'inout'
#plot1.set_xlim(0,5)
#plot1.set_ylim(-2, 14)
plt.axis([0,5, -2, 6])                         #axis[xmin,xmax, ymin,ymax]
#another way to set axis limit
#plt.xlim(xmin,xmax)
#plt.ylim(ymin,ymax)
plt.text(2, 2.5, "...we need 2nd y-scale !!", fontsize = f_size, color = "orange")

legnds = p1+p2+p3
labs = [l.get_label() for l in legnds]
plot1.legend(legnds, labs,  fontsize = f_size-2, loc=0)  #loc is use for location, eg., 0 , "upper left"

#plot1.grid(True)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.savefig("multiyaxisplot0.png")
plt.show()
```

### some options to deal with text file using numpy
https://numpy.org/doc/stable/reference/routines.io.html


- loadtxt(fname[, dtype, comments, delimiter, …])

 Load data from a text file. 
 

- savetxt(fname, X[, fmt, delimiter, newline, …])

 Save an array to a text file.
 

- genfromtxt(fname[, dtype, comments, …])

 Load data from a text file, with missing values handled as specified.


- fromregex(file, regexp, dtype[, encoding])

 Construct an array from a text file, using regular expression parsing.
 

- fromstring(string[, dtype, count, sep])

 A new 1-D array initialized from text data in a string.
 

- ndarray.tofile(fid[, sep, format])

 Write array to a file as text or binary (default).
 

- ndarray.tolist()

 Return the array as an a.ndim-levels deep nested list of Python scalars.
